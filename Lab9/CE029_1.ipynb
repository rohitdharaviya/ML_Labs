**K** means clustering for **iris dataset**.

```python
from sklearn import datasets
iris = datasets.load_iris()
print("Features: ", iris.feature_names)
print("Labels: ", iris.target_names)
```

```python
from sklearn.model_selection import train_test_split
feature=iris.data
target=iris.target
X_train, X_test, y_train, y_test = train_test_split(feature,target, random_state=175)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=38).fit(X_train)
```

```python
 # Get the cluster centroids
print(kmeans.cluster_centers_)
# Get the cluster labels
print(kmeans.labels_)
```

**k** means clustering for **diabetese dataset**.

```python
diabetes = datasets.load_diabetes()
# print(diabetes.data)
data=diabetes.data
print("Features: ", diabetes.feature_names)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=175).fit(data[:,1:3])
```

```python
 # Get the cluster centroids
print(kmeans.cluster_centers_)
# Get the cluster labels
print(kmeans.labels_)
```

```python
# Calculate silhouette_score
from sklearn.metrics import silhouette_score
print(silhouette_score(data[:,1:3], kmeans.labels_))
```

```python
from yellowbrick.cluster import KElbowVisualizer
# Instantiate a scikit-learn K-Means model
model = KMeans(random_state=38)
# Instantiate the KElbowVisualizer with the number of clusters and the metric
visualizer = KElbowVisualizer(model,  k=(2,6), metric='silhouette',timings=False)
# Fit the data and visualize
visualizer.fit(diabetes.data)
visualizer.poof()
```
